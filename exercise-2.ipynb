### Differentiable rendering with a sigmoid function

To have meaningful gradients, we'll use a sigmoid function for the rendering (specifically, the [Logistic function](https://en.wikipedia.org/wiki/Logistic_function)). You'll have to pick the coefficient, which offers a tradeoff between sharpness of the render and informative gradients far from the circle's boundary. Use this interactive plot to explore this tradeoff:

```python
interactive_sigmoid();
```

### Rendering function
Implement the rendering function. It should accept the circle's parameters - center position and radius, and return a 2D tensor of the same dimensions as the ground truth image representing the rendering result.

Use `visualize_render` to inspect your function.

#### Remember: scaling matters

Here you can control the parameters' relative scales.

```python
def render(cx, cy, r):
    x = y = torch.linspace(-10, 10, gt.shape[0])
    xx, yy = torch.meshgrid(x, y, indexing='xy')
    d = torch.sqrt((xx - 3 * cx) ** 2 + (yy - 3 * cy) ** 2)
    return 1 / (1 + torch.exp(-3. * (r - d)))
```

```python
visualize_render(render(0, 0, 5))
```

### Initialization and optimization

Initialize the parameters and optimizer, and implement gradient descent. Some hyper-parameters you can tune:
* Which optimizer to use (e.g. `SGD`, `Adam`)
* Gradient step size ("learning rate", `lr`)
* Number of iterations
* Initialization (though, be careful not to over-rely on initialization in this toy example)

Use the following helper functions to get a visualization of the optimization:
1. `reset_history()` before the optimization
2. `log_history(render, loss)` in every iteration. `render` should be a 2D rendering result, `loss` the scalar value you're minimizing
3. `visualize_history()` after the optimization

```python
cx = torch.tensor(0., requires_grad=True)
cy = torch.tensor(0., requires_grad=True)
radius = torch.tensor(5., requires_grad=True)
optimizer = torch.optim.Adam([cx, cy, radius], lr=1e-1 / 2)
```

```python
reset_history()
for i in range(120):
    result = render(cx, cy, radius)
    diff = ((result - gt) ** 2).mean()

    optimizer.zero_grad()
    diff.backward()
    optimizer.step()

    log_history(result, diff)

visualize_history()
```
